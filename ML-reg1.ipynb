10)What are the key assumptions of Multiple Linear Regression?

Linearity: Relationship between predictors and outcome is linear.

Independence of errors: No correlation between residuals.

Homoscedasticity: Equal variance of residuals.

Normality of residuals: Errors follow normal distribution.

No Multicollinearity: Independent variables should not be highly correlated.

No autocorrelation: Especially important in time series data.


11)What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Definition: Heteroscedasticity occurs when the variance of residuals (errors) is not constant across all levels of the independent variables

Effect:

The model’s coefficient estimates remain unbiased, but the standard errors become unreliable

12)How can you improve a Multiple Linear Regression model with high multicollinearity?

Multicollinearity = strong correlation between independent variables

Solutions:

Remove one of the correlated predictors.

Combine variables (e.g., take average or create an index).

Use Principal Component Analysis (PCA) or dimensionality reduction.

Regularization methods like Ridge Regression or Lasso.

Variance Inflation Factor (VIF) analysis → drop variables with high VIF.

18)What are the limitations of using R² as a sole measure of model performance?

R² always increases when more variables are added (even useless ones).

High R² does not guarantee causation or meaningful predictors.

R² doesn’t show if regression assumptions are violated.

A high R² may still result in poor predictions if overfitting occurs.

Does not account for model complexity (Adjusted R² helps here).

19)How would you interpret a large standard error for a regression coefficient?

Standard Error (SE) measures how precisely a coefficient is estimated.

A large SE means:

Coefficient estimate is unstable/uncertain.

Predictor may not significantly explain variation in Y.

Likely due to small sample size, noise, or multicollinearity.

20)How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Identification:

Plot residuals vs. fitted values.

If variance increases (funnel shape) or decreases systematically → heteroscedasticity.

Importance:

Leads to incorrect standard errors → misleading hypothesis tests.

Affects reliability of confidence intervals & p-values.

Fixes:

Transform dependent variable (e.g., log(Y)).

Use Weighted Least Squares (WLS).

Use Robust Standard Errors.

21)What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

>R² measures the proportion of variance explained by the model.

>Adjusted R² penalizes for adding irrelevant predictors.

> If R² is high but adjusted R² is low, it means the model has too many predictors that don’t add real explanatory power (overfitting).

22)Why is it important to scale variables in Multiple Linear Regression?

Scaling ensures all features are on a comparable scale (e.g., age in years vs. income in lakhs).

Prevents variables with larger ranges from dominating.

Helps with interpretation of coefficients and improves numerical stability of gradient descent algorithms

23)What is polynomial regression?

A type of regression where the relationship between independent and dependent variables is modeled as an nth degree polynomial.

24)How does polynomial regression differ from linear regression?

Linear regression: Fits a straight line (degree 1 polynomial).

Polynomial regression: Fits a curved line (higher-degree polynomial).


> Linear in parameters (coefficients) but non-linear in the predictor variables.

25)When is polynomial regression used?

When data shows a non-linear relationship but can be approximated by a curve.

Example: growth rates, demand curves, disease progression trends.

31)How is polynomial regression implemented in Python?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 6, 14, 28, 45])

# Transform to polynomial features
poly = PolynomialFeatures(degree=2)   # quadratic
X_poly = poly.fit_transform(X)

# Fit model
model = LinearRegression()
model.fit(X_poly, y)

# Predict
y_pred = model.predict(X_poly)

# Visualize
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.show()
```
